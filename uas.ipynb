```python
def histogram(kolom_data):
    non_diabetes = data[data['Diabetes_binary'] == 0][kolom_data].value_counts()
    diabetes = data[data['Diabetes_binary'] == 1][kolom_data].value_counts()
    df = pd.DataFrame([non_diabetes, diabetes])
    df.index = ['Tidak Diabetes', 'Diabetes']
    df.plot(kind = 'bar', stacked = True, figsize = (10,5))
```

### Normalisasi Data

```python
from sklearn import preprocessing

scaler = preprocessing.MinMaxScaler()
d = scaler.fit_transform(data)
data_normalisasi = pd.DataFrame(d, columns = data.columns)
data_normalisasi.head(10)
```

### Menentukan data predictor dan data target

```python
X = data_normalisasi.drop(["Diabetes_binary"], axis = 1)
y = data_normalisasi["Diabetes_binary"]
```

```python
data_normalisasi['Diabetes_binary'].hist()
```

```python
X.head(10)
```

```python
y.head(10)
```

### Split dan Test Data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
y_train.hist()
```

### Balancing Data menggunakan Random Over Sampler

```python
from imblearn.over_sampling import RandomOverSampler
```

```python
X_os, y_os = RandomOverSampler().fit_resample(X_train, y_train)
```

```python
y_os.hist()
```

## Predictive Modeling

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
model_lr = LogisticRegression()
```

```python
##Dengan Over Sampling
model_lr.fit(X_os, y_os)
```

```python
model_lr.score(X_test, y_test)
```

```python
##Tanpa Over Sampling
model_lr.fit(X_train, y_train)
```

```python
model_lr.score(X_test, y_test)
```

### KFold Logistic Regression

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
k_fold = KFold(n_splits = 10, shuffle = True, random_state = 0)
```

```python
score_lr = cross_val_score(model_lr, X, y, cv = k_fold, n_jobs = 1, scoring = 'accuracy')
print(score_lr)
```

```python
score_lr.mean() * 100
```

### K-Nearest NeighBors (KNN)

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
##Dengan Over Sampling
model_knn = KNeighborsClassifier(n_neighbors = 3)
model_knn.fit(X_os, y_os)
```

```python
model_knn.score(X_test, y_test)
```

```python
##Tanpa Over Sampling
model_knn = KNeighborsClassifier(n_neighbors = 3)
model_knn.fit(X_train, y_train)
```

```python
model_knn.score(X_test, y_test)
```

### KFold KNN

```python
score_knn = cross_val_score(model_knn, X, y, cv = k_fold, n_jobs = 1, scoring = 'accuracy')
print(score_knn)
```

```python
score_knn.mean() * 100
```

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model_rf = RandomForestClassifier()
```

```python
##Dengan Over Sampling
model_rf.fit(X_os, y_os)
```

```python
model_rf.score(X_test, y_test)
```

```python
##Tanpa Over Sampling
model_rf.fit(X_train, y_train)
```

```python
model_rf.score(X_test, y_test)
```

### KFold Random Forest

```python
score_rf = cross_val_score(model_rf, X, y, cv = k_fold, n_jobs = 1, scoring = 'accuracy')
print(score_rf)
```
